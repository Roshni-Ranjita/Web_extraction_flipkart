##### Extracting Data

```python
#Extraction Per Page
def Extraction_Per_Page(driver):
    Product_name= driver.find_elements(By.CLASS_NAME,"s1Q9rs") 
    Rating= driver.find_elements(By.CLASS_NAME,"_3LWZlK")
    No_of_People_Rated= driver.find_elements(By.CLASS_NAME,"_2_R_DZ")
    current_cost = driver.find_elements(By.CLASS_NAME,"_30jeq3") 
    Original_Cost= driver.find_elements(By.CLASS_NAME,"_3I9_wc") 
    Percentage_off= driver.find_elements(By.CLASS_NAME,"_3Ay6Sb") 
    df=pd.DataFrame(columns=['Product_name', 'Rating', 'No_of_People_Rated','current_cost','Original_Cost','Percentage_off'])
    for na,r,no,cc,oc,pt in zip(Product_name, Rating,No_of_People_Rated,current_cost,Original_Cost,Percentage_off) :
        df = df.append({'Product_name':na.text, 'Rating':r.text ,'No_of_People_Rated':no.text, 'current_cost':cc.text, 'Original_Cost':oc.text, 'Percentage_off':pt.text},ignore_index=True)
    return(df)
```
